```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.head(15)
```

```python
data['age'].fillna(data['age'].mean(), inplace=True)
```

```python
data.head(15)
```

```python
data['deck'].fillna(method='ffill', inplace=True)
```

```python
data['deck'].fillna(method='bfill', inplace=True)
```

```python
data.head(8)
```

```python
data.dropna(subset=['embarked'], inplace=True)
```

```python
data.dropna(subset=['embark_town'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data['sex'] = pd.factorize(data['sex'])[0]
```

```python
data['embarked'] = pd.factorize(data['embarked'])[0]
```

```python
data['class'] = pd.factorize(data['class'])[0]
```

```python
data['who'] = pd.factorize(data['who'])[0]
```

```python
data['adult_male'] = pd.factorize(data['adult_male'])[0]
```

```python
data['deck'] = pd.factorize(data['deck'])[0]
```

```python
data['embark_town'] = pd.factorize(data['embark_town'])[0]
```

```python
data['alive'] = pd.factorize(data['alive'])[0]
```

```python
data['alone'] = pd.factorize(data['alone'])[0]
```

```python
data['age'] = (data['age'] - data['age'].min()) / (data['age'].max() - data['age'].min())
```

```python
data['fare'] = (data['fare'] - data['fare'].min()) / (data['fare'].max() - data['fare'].min())
```

```python
data['pclass'] = (data['pclass'] - data['pclass'].min()) / (data['pclass'].max() - data['pclass'].min())
```

```python
data.head(4)
```
